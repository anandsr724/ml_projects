```python
c = df["flight"].value_counts()
index_less_than_10 = c[c<=10].index
index_less_than_100 = c[(c<=100) & (c>10)].index
index_less_than_500 = c[(c<=500) & (c>100)].index
index_greater_than_500 = c[c>500].index
```

```python
# to_replace=df["location"].value_counts()[df["location"].value_counts()<10].index

# df['location'].replace(to_replace, 'others',inplace=True)

df['flight'].replace(index_less_than_10, 1,inplace=True)
df['flight'].replace(index_less_than_100, 2,inplace=True)
df['flight'].replace(index_less_than_500, 3,inplace=True)
df['flight'].replace(index_greater_than_500, 4,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X=df.drop("price",axis=1)
y=df["price"]
X_train , X_test , y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
X_train.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import StandardScaler


#flight -> ohe with others etc

# normal ohe -> source_city , destination_city  , airline   , class

# ordinal encoding -> departure_time , arrival_time , stops 
transformer = ColumnTransformer(
    transformers= [ 
    ("normal_ohe" , OneHotEncoder(sparse_output=False,drop='first'),["source_city","destination_city","airline","class"]),
    ("ordinal_atime" , OrdinalEncoder(categories=[["Early_Morning","Morning","Afternoon","Evening","Night","Late_Night"]]),["arrival_time"]),
    ("ordinal_dtime" , OrdinalEncoder(categories=[["Early_Morning","Morning","Afternoon","Evening","Night","Late_Night"]]),["departure_time"]),
    ("ordinal_stops" , OrdinalEncoder(categories=[["zero","one","two_or_more"]]),["stops"]),
    # ("scalar",StandardScaler(),),
    

    # ("ordinal_encoding" , KBinsDiscretizer(n_bins=10,encode='ordinal',strategy='quantile'),["stops"]),

    ], remainder='passthrough'
)
```

```python
transformer.fit(X_train)
```

```python
X_train_trf=transformer.transform(X_train)
X_test_trf=transformer.transform(X_test)
```

```python
X_train_trf.shape
```

```python
X_test_trf.shape
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()


lr.fit(X_train_trf,y_train)
y_pred=lr.predict(X_test_trf)

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

print("RMSE : ", mean_squared_error(y_test,y_pred))

print("MAE : ", mean_absolute_error(y_test,y_pred))

print("R2 Score : ", r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor()


rf.fit(X_train_trf,y_train)
y_pred=rf.predict(X_test_trf)

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

print("RMSE : ", mean_squared_error(y_test,y_pred))

print("MAE : ", mean_absolute_error(y_test,y_pred))

print("R2 Score : ", r2_score(y_test,y_pred))
```
